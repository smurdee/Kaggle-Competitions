```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df = pd.concat(objs = [df, pd.read_csv('test.csv')], axis = 0, ignore_index = True)
df.head(20)
```

```python
df.tail(20)
```

```python
df = df.drop('PassengerId', axis=1)
df.head(20)
```

```python
df = pd.get_dummies(df, columns = ['Pclass', 'Embarked'], dtype='int64')
df.head(20)
```

```python
df[['Last_Name', 'First_Name']] = df.Name.str.split(',', expand=True)
df = df.drop('Name', axis=1)
df.head(20)
```

```python
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
df = pd.get_dummies(df, columns = ['Cabin', 'Last_Name'], dtype='int64')
df.head(20)
```

```python
df.iloc[0:len(pd.read_csv('train.csv')),].to_csv('train_processed.csv', index=False)
```

```python
df.iloc[len(pd.read_csv('train.csv')):,].drop('Survived', axis = 1).to_csv('test_processed.csv', index=False)
```

```python
len(pd.read_csv('train_processed.csv')) == len(pd.read_csv('train.csv'))
```

```python
len(pd.read_csv('test_processed.csv')) == len(pd.read_csv('test.csv'))
```
